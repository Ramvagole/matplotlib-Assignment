```python
#Q2):-
"""The scatter diagram graphs numerical data pairs, with one variable on each axis, show their relationship."""
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
#Q3):-
"""The acts as a utility wrapper and helps in creating common layouts of subplots, including the enclosing figure object, in a single call."""

import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(4,1,1)
plt.plot(x,y)
plt.title("First figure")

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(4,1,2)
plt.plot(x,y)
plt.title("Second figure")

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(4,1,3)
plt.plot(x,y)
plt.title("Third figure")

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(4,1,4)
plt.plot(x,y)
plt.title("Fourth figure")

plt.tight_layout()
plt.show()
```

```python
#Q4):-
"""when you want to show a distribution of data points or perform a comparison of metric values across different subgroups of your data."""

import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.show()
```

```python
#Q5):-
"""Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups."""

import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.show()
```
